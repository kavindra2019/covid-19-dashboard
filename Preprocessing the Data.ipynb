```python
import pandas as pd

df=pd.read_csv('covid19.csv')
df.head()
```

```python
df=df.loc[:,['Country_Region','Province_State','Last_Update','Copy_of_Confirmed','Copy_of_Deaths','Copy_of_Recovered']]
df.head()
```

```python
df['Last_Update']=pd.to_datetime(df['Last_Update']).dt.date
df.head()
```

```python
df_world=df.groupby('Last_Update').agg({'Copy_of_Confirmed':['sum'],'Copy_of_Deaths':['sum'],'Copy_of_Recovered':['sum']})
df_world.reset_index(inplace=True)
df_world=df_world.sort_values('Last_Update')
df_world.head()
```

```python
df_world['Copy_of_Recovered']=pd.to_numeric(df_world['Copy_of_Recovered'].iloc[:,0], downcast='integer')
def to_daily():
    df_world['Daily_Confirmed'] = df_world['Copy_of_Confirmed'].diff().fillna(df_world['Copy_of_Confirmed'])
    df_world['Daily_Deaths'] = df_world['Copy_of_Deaths'].diff().fillna(df_world['Copy_of_Deaths'])
    df_world['Daily_Recovered'] = df_world['Copy_of_Recovered'].diff().fillna(df_world['Copy_of_Recovered'])
to_daily()
df_world.iloc[:,:]
```

```python
def df_world_drop(arr):
    df_world.drop(arr,axis=0,inplace=True)
    df_world.reset_index(inplace=True,drop=True)
drop_indexes = [23,24,25,28,29,30,31,32,33,34,46,47,48,57,58,59,60,61,62,63,64,65,66,67,68,88,41,42,43,44,45,56]
df_world_drop(drop_indexes)
df_world.iloc[30:,:]
```
